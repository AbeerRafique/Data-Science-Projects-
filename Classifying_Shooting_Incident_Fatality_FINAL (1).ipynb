## **1.1 Project Overview**

This project is a collaboration between our data science team and Urban Insights Analytics, a leading data analytics firm specializing in urban crime analysis. The objective is to enhance emergency response strategies and optimize resource deployment across New York City by analyzing historical shooting incident data. Using machine learning techniques, the project aims to classify shooting incidents as either fatal or non-fatal, ultimately contributing to the development of more informed and proactive public safety measures.

## **1.2 Introduction**

Gun violence remains a critical public safety concern in major urban areas, particularly in cities like New York. Understanding the factors that contribute to the fatality of shooting incidents can play a pivotal role in improving emergency response planning and reducing future incidents. With increasing access to historical crime data, there is a unique opportunity to leverage machine learning to identify patterns and risk factors that may influence fatal outcomes in shooting cases.

This project focuses on building a predictive model that classifies whether a shooting incident results in a fatality (`STATISTICAL_MURDER_FLAG: TRUE`) or not (`FALSE`). The insights generated from this model will support targeted interventions, more efficient allocation of resources, and better-informed policing strategies.

## **1.3 Problem Statement**

The primary challenge is to build a reliable classification model that can predict the likelihood of a shooting incident being fatal, based on historical data. Accurately classifying fatal versus non-fatal incidents is essential for prioritizing law enforcement and emergency response efforts, as well as for understanding the underlying socio-demographic and situational factors that elevate the risk of fatality.

- To develop a machine learning model capable of accurately classifying shooting incidents as fatal or non-fatal.

# **5. Data Preprocessing & Cleaning**

In this section, we clean, structure, and transform the raw NYC shooting incident data to prepare it for predictive modeling. The preprocessing steps include dropping irrelevant or sparse columns, standardizing null values, feature engineering on temporal data, imputing missing values, encoding categorical features, outlier removal, and creating a final model-ready dataset.

```python
# Copy original to avoid modifying in-place
df = train_df.copy()
```

## **5.1 Parse Date and Time Columns**

```python
# Parse OCCUR_DATE and OCCUR_TIME
df['OCCUR_DATE'] = pd.to_datetime(df['OCCUR_DATE'], errors='coerce')
df['OCCUR_TIME'] = pd.to_datetime(df['OCCUR_TIME'], format='%H:%M:%S', errors='coerce').dt.time
```

Purpose: Convert string-formatted dates and times into datetime objects for downstream temporal feature engineering.

## **5.2 Normalize Null-Like String Values**

```python
# Replace common null-like strings with np.nan
null_like_values = ['(null)', 'NULL', 'Null', 'null', 'unknown', 'UNKNOWN', 'Unk', 'UNK']

cols_with_cats = df.select_dtypes(include='object').columns

for col in cols_with_cats:
    df[col] = df[col].replace(null_like_values, np.nan)
```

Purpose: Standardize null representations ('null', 'unknown', etc.) into proper np.nan for reliable missing value handling.

## **5.3 Drop High-Null and Non-Essential Columns**

```python
# Drop LOC_OF_OCCUR_DESC and LOC_CLASSFCTN_DESC (both ~94% missing)
df.drop(['LOC_OF_OCCUR_DESC', 'LOC_CLASSFCTN_DESC'], axis=1, inplace=True)
```

Purpose: Remove columns with excessive missing data or redundant/irrelevant information.

## **5.4  Handle Critical Missing Data**

```python
# Drop rows where PERP_AGE_GROUP, PERP_SEX, PERP_RACE are missing
df = df.dropna(subset=['PERP_AGE_GROUP', 'PERP_SEX', 'PERP_RACE'])

# Fill LOCATION_DESC with 'UNKNOWN' for modeling if needed
df['LOCATION_DESC'] = df['LOCATION_DESC'].fillna('UNKNOWN')

# Drop remaining rows with missing coordinates (only 10)
df = df.dropna(subset=['Latitude', 'Longitude', 'X_COORD_CD', 'Y_COORD_CD'])

# If Lon_Lat can be reconstructed from Latitude & Longitude, drop it
df.drop('Lon_Lat', axis=1, inplace=True)
```

Purpose: Remove or impute records with missing values in essential features to maintain data integrity before modeling

## **5.5 Fix Data Types**

```python
# Ensure correct types
df['JURISDICTION_CODE'] = df['JURISDICTION_CODE'].astype('Int64')  # allows NA
df['PRECINCT'] = df['PRECINCT'].astype('int')
df['STATISTICAL_MURDER_FLAG'] = df['STATISTICAL_MURDER_FLAG'].astype(bool)
```

Purpose: Set appropriate data types for modeling and memory efficiency.

## **5.6 Missing Value Diagnostics**

```python
# Final checks
print(df.isnull().sum())
print(df.dtypes)
print(df.shape)
```

Purpose: Review remaining missing values and data types before imputations.

## **5.7 Impute Missing Values**

```python
# Impute JURISDICTION_CODE with mode
df['JURISDICTION_CODE'] = df['JURISDICTION_CODE'].fillna(df['JURISDICTION_CODE'].mode()[0])

# Impute VIC_AGE_GROUP with "Unknown" (common for categories)
df['VIC_AGE_GROUP'] = df['VIC_AGE_GROUP'].fillna('UNKNOWN')
```

Purpose: Fill remaining missing values using domain-specific logic.

## **5.8  Temporal Feature Extraction**

```python
df['OCCUR_YEAR'] = df['OCCUR_DATE'].dt.year
df['OCCUR_MONTH'] = df['OCCUR_DATE'].dt.month
df['OCCUR_DAY'] = df['OCCUR_DATE'].dt.day
df['OCCUR_HOUR'] = pd.to_datetime(df['OCCUR_TIME'], errors='coerce').dt.hour

df['SEASON'] = (df['OCCUR_MONTH'] % 12 // 3 + 1).map({1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Fall'})
```

Purpose: Derive calendar and seasonal features from the incident date and time.

## **5.9  Refine Time Features**

```python
# Convert OCCUR_TIME
df['OCCUR_TIME'] = pd.to_datetime(df['OCCUR_TIME'], format='%H:%M:%S', errors='coerce')

# Extract Time-Based Features
df['HOUR'] = df['OCCUR_TIME'].dt.hour
df['TIME_OF_DAY'] = pd.cut(df['HOUR'],
    bins=[-1, 5, 11, 17, 21, 24],
    labels=['Night', 'Morning', 'Afternoon', 'Evening', 'Late Night']
)
```

Purpose: Re-parse OCCUR_TIME and categorize hour into meaningful time periods.

## **5.10 Deduplicate Incident Records**

```python
# Check for duplicate incident rows
duplicates = df.duplicated(subset='INCIDENT_KEY').sum()
print(f"Duplicate INCIDENT_KEYs: {duplicates}")
```

```python
df = df.drop_duplicates(subset='INCIDENT_KEY', keep='first')
print(f"Remaining duplicates: {df.duplicated(subset='INCIDENT_KEY').sum()}")
print(f"New shape: {df.shape}")
```

Purpose: Ensure uniqueness of INCIDENT_KEY to avoid duplicated rows.

## **9.1: Replace Rare Jurisdiction Code**

```python
# Replace rare jurisdiction codes with a single 'other' category
jurisdiction_freq = df['JURISDICTION_CODE'].value_counts(normalize=True)
rare_codes = jurisdiction_freq[jurisdiction_freq < 0.01].index  # Less than 1%
df['JURISDICTION_CODE'] = df['JURISDICTION_CODE'].replace(rare_codes, 999)  # 999 = 'Other'
```

Purpose:
To simplify categorical encoding and reduce noise by grouping infrequent jurisdiction codes (less than 1% frequency) into a single "Other" category.

## **9.2: Define Outlier Capping Function Using IQR**

```python
def cap_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] < lower, lower,
                       np.where(df[col] > upper, upper, df[col]))
    return df

# Apply to both coordinates
df = cap_outliers_iqr(df, 'X_COORD_CD')
df = cap_outliers_iqr(df, 'Longitude')
```

Purpose:
To reduce the influence of extreme outliers by capping values outside the IQR bounds (1.5 * IQR) and To limit outliers specifically in geospatial features like X_COORD_CD and Longitude.

## **11.3 Threshold Tuning for Balanced Random Forest Classifier**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, roc_auc_score, average_precision_score, f1_score, precision_score, recall_score # Import precision_score and recall_score

# Probabilities from your model (adjust variable if needed)
probs = y_proba_rf  # RF predicted probabilities for class 1 (Fatal)
thresholds = np.arange(0.25, 0.45, 0.01)

precisions, recalls, f1s = [], [], []

for t in thresholds:
    preds = (probs >= t).astype(int)
    p = precision_score(y_test, preds)
    r = recall_score(y_test, preds)
    f1 = f1_score(y_test, preds)
    precisions.append(p)
    recalls.append(r)
    f1s.append(f1)

# Plotting
plt.figure(figsize=(10,6))
plt.plot(thresholds, precisions, label='Precision', marker='o')
plt.plot(thresholds, recalls, label='Recall', marker='o')
plt.plot(thresholds, f1s, label='F1 Score', marker='o')
plt.axvline(x=0.31, color='gray', linestyle='--', label='Threshold = 0.31')
plt.title('Precision, Recall, and F1 Score vs. Threshold (Random Forest)')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**This graph shows how good a "Random Forest" computer model is at making predictions, depending on a "threshold" setting.**

- Precision (blue line): How accurate the model is when it says something is true. Higher is better.

- Recall (orange line): How many of the actual true things the model correctly finds. Higher is better.

- F1 Score (green line): A balance between precision and recall. It gives you an idea of overall performance.

**What the graph tells us:**

- Trade-off: As the "threshold" increases (moving right on the bottom axis), the model becomes more picky. It gets better at being accurate when it predicts something (precision goes up), but it misses more of the actual true things (recall goes down).

- The dashed line at "Threshold = 0.31": At this specific setting, the model is very good at finding most of the true things (high recall, around 82%). However, when it does say something is true, it's not very accurate (low precision, around 23%). The overall balance (F1 Score) at this point is around 36%.

In short: At a threshold of 0.31, this model is good at catching almost all of what it's looking for, but it also makes a lot of incorrect "true" predictions.

## **11.4 Precision, Recall, and F1 vs. Threshold Curve (Random Forest)**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve

# Step 1: Get precision, recall, thresholds from predicted probabilities
precision, recall, thresholds = precision_recall_curve(y_test, y_proba_rf)

# Step 2: Calculate F1 scores safely
f1_scores = 2 * (precision * recall) / (precision + recall + 1e-6)  # Add epsilon to avoid divide-by-zero

# Step 3: Plot the curves
plt.figure(figsize=(10, 6))
plt.plot(thresholds, precision[:-1], label='Precision', color='blue')
plt.plot(thresholds, recall[:-1], label='Recall', color='green')
plt.plot(thresholds, f1_scores[:-1], label='F1 Score', color='red')
plt.axvline(x=0.35, color='gray', linestyle='--', label='Current Threshold = 0.31')

plt.title('Precision, Recall, and F1 Score vs Threshold (Random Forest)')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Interpretation at the Current Threshold (0.31)**:

At this chosen threshold, the Random Forest model exhibits:

- High Recall: It successfully identifies about 80% of the actual positive cases. This is good if missing positive cases is costly.

- Low Precision: However, when the model predicts something is positive, it's only correct about 22% of the time. This means it has a significant number of false positives.

- Moderate F1 Score: The F1 score of 0.35 reflects this trade-off. While it's good at catching positives (recall), its overall accuracy for those predictions (precision) is quite low.

In summary: The graph illustrates the inherent trade-off in machine learning models between precision and recall. The specific threshold chosen (0.31 in this case) prioritizes recall over precision, leading to a model that is good at finding most positive instances but also produces many false alarms.

## **Threshold Selection Strategy for Random Forest: Prioritizing Recall in Critical Case Detection.**


For a model designed to detect **fatal and non-fatal cases**, prioritizing recall at a threshold of 0.31 is a **critically sound and highly justifiable strategic decision.**

Here is why selecting 0.31 and emphasizing recall is advantageous for the model's application:

1.  **Minimizing Catastrophic Omissions (Fatal Cases):**
    * **The Problem:** The failure to identify a "fatal case" (a false negative concerning a fatal outcome) carries the gravest of consequences, potentially leading to loss of life or severe, irreversible harm.
    * **The Solution (High Recall):** By setting the threshold at 0.31, the model achieves a high recall (approximately 80%). This design choice ensures the model is highly effective at **identifying nearly all actual fatal and non-fatal cases.** This significantly reduces the risk of a fatal case going undetected.
    * **The Advantage:** This approach prioritizes patient safety and the prevention of the most severe outcomes. High recall ensures that very few *actual* serious situations are overlooked, thereby maximizing the opportunity for timely intervention.

2.  **Enabling Early Warning and Intervention (Both Fatal & Non-Fatal):**
    * **The Problem:** Delays in identifying any case, even if ultimately non-fatal, could potentially exacerbate its severity or complicate its management.
    * **The Solution (High Recall):** The model is configured to flag a broad spectrum of potential issues. This provides an early alert for a substantial percentage of *all* cases requiring attention.
    * **The Advantage:** Even if a flagged case proves to be "non-fatal" (which could be a false positive if the sole focus is fatal outcomes, or a true positive if the "non-fatal" condition itself requires attention), the early notification facilitates prompt assessment and potentially prevents a non-fatal case from worsening, or ensures timely management for non-fatal but still critical conditions.

3.  **Balancing Manageable False Alarms Against Unacceptable Misses:**
    * **The Problem:** At a threshold of 0.31, the model's precision is lower (around 22%). This implies that a considerable number of flagged cases might not be fatal, or perhaps not even true cases in a strict sense (depending on class definitions and if "non-fatal" is also a positive class). These represent the "false alarms" when the focus is primarily on fatal outcomes.
    * **The Solution (Accepting Lower Precision):** A deliberate decision has been made that the effort required to investigate these "false alarms" (e.g., through further diagnostic tests, medical review, or follow-up) is considerably less detrimental than the catastrophic repercussions of failing to detect an actual fatal case.
    * **The Advantage:** It is demonstrably more feasible and less impactful to review and ultimately dismiss a few additional "non-fatal" flags than to contend with the aftermath of a missed fatal diagnosis. The resources allocated to investigating false positives are viewed as a necessary investment in averting genuine tragedies.

**In essence, by opting for a threshold of 0.31, the model functions as an exceptionally sensitive early warning system. This choice signifies that it is of paramount importance to *not miss* a fatal case, and that a willingness to investigate more less severe or even non-existent issues is acceptable to establish that crucial safety net.** This strategy is directly aligned with the severe ramifications associated with the failure to detect fatal conditions.

# **13. Limitations, Recommendations, and Future Work**


## **13.1 Limitations**

1. **Class Imbalance Challenge**
   The dataset is highly imbalanced, with a much smaller proportion of fatal cases. While techniques like SMOTE and threshold tuning help, the model still suffers from lower precision due to the scarcity of positive (fatal) samples.

2. **Limited Feature Granularity**
   Some potentially influential features (e.g., type of weapon, mental state of the shooter, location context, police involvement) were not present in the dataset. This limits the model's ability to fully capture situational risk factors.

3. **Temporal & Spatial Drift**
   The model does not currently account for temporal trends (e.g., seasonal crime patterns) or spatial autocorrelation (e.g., clustering of shootings in specific precincts). This may reduce generalizability to future or unseen incidents.

4. **Threshold Trade-off**
   The chosen threshold (0.35) offers improved recall but sacrifices precision, which may result in an increased number of false positivespotentially leading to resource misallocation if used operationally without human review.

5. **Static Model Deployment**
   The model is trained on historical data but is not designed to dynamically update in real-time with new incident data. This introduces a risk of model staleness in evolving urban environments.


### **13.2 Recommendations**

1. **Integrate Domain Expertise**
   Collaborate with criminologists, sociologists, and public safety experts to design more relevant features and contextual interpretations of model predictions.

2. **Precision-Oriented Post-Processing**
   Use a two-stage approach: the model flags potential fatal cases, and a second-level decision system or human analyst verifies high-risk predictions before escalation.

3. **Incident Context Enrichment**
   Augment data with external sources such as:

   * Local crime trends
   * Historical violence rates by location
   * Real-time 911 or social media feeds
     to improve context-awareness of each case.

4. **Monitor Model Drift**
   Implement model monitoring pipelines that check for distributional drift in key features and prediction patterns over time.

## **13.3 Future Work**
1. **Spatio-Temporal Modeling**
   Incorporate geospatial techniques (e.g., hotspot mapping, spatial lag models) and time-series modeling (e.g., LSTM, ARIMA) to improve prediction accuracy over time and location.

2. **Survival Analysis & Risk Scoring**
   Shift from binary classification to **probabilistic survival models** or **risk scoring frameworks** that estimate the likelihood of escalation to fatality based on progressive incident details.

3. **Real-Time Pipeline & Alerting System**
   Build a full-stack pipeline that ingests real-time data (e.g., from police records or IoT crime detection systems), performs preprocessing and inference, and sends alerts to relevant authorities based on model confidence.

4. **Causal Inference for Intervention**
   Extend the framework to answer **what-if questions** using causal models to simulate the effect of interventions like increased patrol or policy change on fatality reduction.

5. **Multimodal Data Fusion**
   Explore combining structured data with unstructured sources (e.g., crime narratives, emergency call transcripts, CCTV summaries) to extract deeper signals for fatality prediction.
