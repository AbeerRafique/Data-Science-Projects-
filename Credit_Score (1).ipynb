# **1. Project Overview**
This project focuses on building an automated credit score classification system using machine learning to enhance the credit assessment process in financial institutions. By analyzing a rich dataset of personal and financial attributes, the goal is to classify individuals into predefined credit score categories (e.g., Good, Average, Poor) efficiently and accurately. The workflow includes exploratory data analysis, data preprocessing, feature engineering, and model evaluation using algorithms like Random Forest and XGBoost. The project emphasizes not only predictive accuracy but also interpretability, ensuring that important factors influencing credit decisions are clearly identified. Ultimately, the system aims to reduce manual workload, minimize human bias, and support faster, data-driven credit decisions for improved customer experience and risk management.

# **2. Introduction**
Credit scores are a key part of financial systems. They help banks and lenders decide whether a person is likely to repay loans or not. Traditionally, credit score assessment is done manually, which can take time and may lead to inconsistent decisions. By using machine learning, we can automate this process and make it faster, more accurate, and fairer. This project aims to build a machine learning model that can classify a person’s credit score (such as Good, Average, or Poor) based on their financial and personal data.

# **3. Problem Statement**
Assessing a person's creditworthiness manually is time-consuming and prone to human error. Inconsistent or delayed credit decisions can negatively affect both lenders and customers. There is a need for a reliable, automated system that can classify credit scores efficiently using data-driven methods. Without such a system, financial institutions risk giving loans to high-risk individuals or rejecting trustworthy applicants.

## **8.3 Missing Value Treatment and Column Cleaning**

````python
# 1. Monthly_Inhand_Salary (float)
train_df['Monthly_Inhand_Salary'] = train_df['Monthly_Inhand_Salary'].fillna(train_df['Monthly_Inhand_Salary'].median())

# 2. Type_of_Loan (multi-loan object column)
train_df['Type_of_Loan'] = train_df['Type_of_Loan'].fillna("Not Specified")

# 3. Name (drop if exists)
if 'Name' in train_df.columns:
    train_df.drop(columns=['Name'], inplace=True)

# 4. Num_of_Delayed_Payment (object → numeric → impute)
train_df['Num_of_Delayed_Payment'] = pd.to_numeric(train_df['Num_of_Delayed_Payment'], errors='coerce')
train_df['Num_of_Delayed_Payment'] = train_df['Num_of_Delayed_Payment'].fillna(train_df['Num_of_Delayed_Payment'].median())

# 5. Amount_invested_monthly (object → numeric → impute)
train_df['Amount_invested_monthly'] = pd.to_numeric(train_df['Amount_invested_monthly'], errors='coerce')
train_df['Amount_invested_monthly'] = train_df['Amount_invested_monthly'].fillna(train_df['Amount_invested_monthly'].median())

# 6. Num_Credit_Inquiries (float → impute)
train_df['Num_Credit_Inquiries'] = train_df['Num_Credit_Inquiries'].fillna(train_df['Num_Credit_Inquiries'].median())

# 7. Monthly_Balance (object → numeric → impute)
train_df['Monthly_Balance'] = pd.to_numeric(train_df['Monthly_Balance'], errors='coerce')
train_df['Monthly_Balance'] = train_df['Monthly_Balance'].fillna(train_df['Monthly_Balance'].median())


# 8. Drop fully missing column
if 'Payment_of_Min_Amount' in train_df.columns and train_df['Payment_of_Min_Amount'].isnull().sum() == len(train_df):
    train_df.drop(columns=['Payment_of_Min_Amount'], inplace=True)

# 9. Credit_History_Age (string → numeric)
if 'Credit_History_Age' in train_df.columns:
    # Extract years and months, coercing errors to NaN
    extracted = train_df['Credit_History_Age'].str.extract(r'(\d+)\s+Years.*?(\d+)\s+Month')
    train_df['Credit_History_Years'] = pd.to_numeric(extracted[0], errors='coerce')
    train_df['Credit_History_Months'] = pd.to_numeric(extracted[1], errors='coerce')

    # Calculate total months and fill NaNs
    train_df['Credit_History_Total_Months'] = train_df['Credit_History_Years'] * 12 + train_df['Credit_History_Months']
    train_df['Credit_History_Total_Months'] = train_df['Credit_History_Total_Months'].fillna(train_df['Credit_History_Total_Months'].median())

    # Drop intermediate and original column
    train_df.drop(['Credit_History_Age', 'Credit_History_Years', 'Credit_History_Months'], axis=1, inplace=True)


# Final check: zero missing values
print("Remaining missing values:\n", train_df.isnull().sum())
````

## **9.3 Detecting Outliers Using the IQR Method**

````python
# Create a copy to avoid modifying the original
iqr_outliers = pd.DataFrame(index=train_df.index)

for col in num_cols:
    # Check if the column is numeric before proceeding
    if pd.api.types.is_numeric_dtype(train_df[col]):
        Q1 = train_df[col].quantile(0.25)
        Q3 = train_df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Flag outliers (1 if outlier, 0 if not)
        iqr_outliers[col + '_outlier'] = ((train_df[col] < lower_bound) | (train_df[col] > upper_bound)).astype(int)
    else:
        print(f"Skipping non-numeric column: {col}")


# See how many outliers per column
outlier_counts = iqr_outliers.sum().sort_values(ascending=False)
print("Number of outliers detected per feature:")
print(outlier_counts)
````

**IQR-Based Outlier Detection**

Applies the Interquartile Range (IQR) method to flag outliers in numerical features:

- For each numeric column:
  - Computes Q1 (25th percentile), Q3 (75th percentile), and IQR (Q3 - Q1)
  - Flags values below `Q1 - 1.5*IQR` or above `Q3 + 1.5*IQR` as outliers
  - Creates binary outlier flags per column (`1 = outlier`, `0 = normal`)

The resulting `iqr_outliers` DataFrame summarizes outlier presence for each feature, helping prioritize columns for outlier treatment.

The final output (`outlier_counts`) shows which features have the most extreme values requiring review.

# **10. Outliers Treatment**

## **10.1** Winsorize (Cap) the Outliers

````python
# Store IQR bounds
iqr_bounds = {}

df_capped_train = train_df.copy()

for col in num_cols:
    # Check if the column is numeric before proceeding
    if pd.api.types.is_numeric_dtype(train_df[col]):
        Q1 = train_df[col].quantile(0.25)
        Q3 = train_df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        iqr_bounds[col] = (lower_bound, upper_bound)

        df_capped_train[col] = np.where(train_df[col] < lower_bound, lower_bound,
                             np.where(train_df[col] > upper_bound, upper_bound, train_df[col]))
    else:
        print(f"Skipping non-numeric column for capping: {col}")
````

**Outlier Treatment via IQR Capping**

Caps extreme values in numeric columns using the IQR method to reduce the effect of outliers without removing data points.

Steps:
- Computes IQR bounds for each numeric column
- Stores `(lower_bound, upper_bound)` in a dictionary `iqr_bounds`
- Replaces values below or above the bounds with the respective threshold

Note:
- Non-numeric column `Num_of_Loan` was skipped as it's improperly typed and needs separate preprocessing.
- Output dataset `df_capped_train` is a cleaned version with capped outliers, ready for scaling and modeling.

## 16.3 **RandomForest Classifier**

````python
# Step 1: Imports
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Step 2: Encode target labels
label_mapping = {'Poor': 0, 'Standard': 1, 'Good': 2}
y_train_encoded = y_train.map(label_mapping)
y_test_encoded = y_test.map(label_mapping)

# Step 3: Handle NaNs and Infs in X_train and X_test
X_train_clean = X_train_numeric.replace([np.inf, -np.inf], np.nan)
X_train_clean.fillna(X_train_clean.mean(), inplace=True)

X_test_clean = X_test_numeric.replace([np.inf, -np.inf], np.nan)
X_test_clean.fillna(X_test_clean.mean(), inplace=True)

# Step 4: Initialize and Train Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_clean, y_train_encoded)

# Step 5: Predict
rf_y_pred = rf_model.predict(X_test_clean)

# Step 6: Accuracy
print(" Random Forest Accuracy:", accuracy_score(y_test_encoded, rf_y_pred))

# Step 7: Classification Report
print("\n Classification Report:\n")
print(classification_report(y_test_encoded, rf_y_pred, target_names=label_mapping.keys()))

# Step 8: Confusion Matrix
cm = confusion_matrix(y_test_encoded, rf_y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=label_mapping.keys(),
            yticklabels=label_mapping.keys())
plt.title("Random Forest Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
````

## **20.3 Detecting Outliers Using the IQR Method**

````python
# Create a copy to avoid modifying the original test_df
iqr_outliers_test = pd.DataFrame(index=test_df.index)

for col in num_cols:
    # Ensure the column exists and is numeric
    if col in test_df.columns and pd.api.types.is_numeric_dtype(test_df[col]):
        # Use train_df quantiles to ensure consistency
        Q1 = train_df[col].quantile(0.25)
        Q3 = train_df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Flag outliers (1 if outlier, 0 if not)
        iqr_outliers_test[col + '_outlier'] = ((test_df[col] < lower_bound) | (test_df[col] > upper_bound)).astype(int)
    else:
        print(f"Skipping non-numeric or missing column: {col}")

# Count of outliers per column in test set
outlier_counts_test = iqr_outliers_test.sum().sort_values(ascending=False)
print("Number of outliers detected per feature in test_df:")
print(outlier_counts_test)
````

A large number of outliers were flagged in nearly every numeric feature. This insight helps decide:
- Whether to cap/floor extreme values
- Whether to use robust models less sensitive to outliers (e.g., tree-based models)
- Whether to use transformations or scaling

The final handling decision depends on model sensitivity and business needs.

## **20.4 Outliers Treatment**

````python
# Apply IQR-based capping to test_df using bounds from train_df
df_capped_test = test_df.copy()

for col in num_cols:
    if col in df_capped_test.columns and pd.api.types.is_numeric_dtype(df_capped_test[col]):
        if col in iqr_bounds:
            lower_bound, upper_bound = iqr_bounds[col]
            df_capped_test[col] = np.where(df_capped_test[col] < lower_bound, lower_bound,
                                  np.where(df_capped_test[col] > upper_bound, upper_bound, df_capped_test[col]))
        else:
            print(f"No IQR bounds found for: {col}")
    else:
        print(f"Skipping non-numeric or missing column for capping: {col}")
````

**Apply IQR-Based Outlier Capping to Test Set**

I applied **IQR-based capping** to the `test_df` using bounds previously calculated from the training set.

- **Objective**: Prevent the model from being affected by extreme values in the test data.
- For each numeric column:
  - Values below the lower bound are capped to the lower bound.
  - Values above the upper bound are capped to the upper bound.
- Only columns present in both `test_df` and `iqr_bounds` are processed.
- Non-numeric or missing columns are automatically skipped.

This step ensures consistency in feature scaling and outlier treatment between training and testing datasets.

## **21. Feature Engineering**

````python
# 1. Debt to Income Ratio
test_df["Debt_Income_Ratio"] = test_df["Outstanding_Debt"] / (test_df["Annual_Income"] + 1)

# 2. EMI to Salary Ratio
test_df["EMI_Salary_Ratio"] = test_df["Total_EMI_per_month"] / (test_df["Monthly_Inhand_Salary"] + 1)

# 3. Convert loan/account columns to numeric
test_df["Num_of_Loan"] = pd.to_numeric(test_df["Num_of_Loan"], errors='coerce').fillna(0)
test_df["Num_Bank_Accounts"] = pd.to_numeric(test_df["Num_Bank_Accounts"], errors='coerce').fillna(0)

# 4. Loans per Account
test_df["Loans_per_Account"] = test_df["Num_of_Loan"] / (test_df["Num_Bank_Accounts"] + 1)

# 5. Utilization per Limit Change
test_df["Utilization_per_LimitChange"] = test_df["Credit_Utilization_Ratio"] / (test_df["Changed_Credit_Limit"] + 1)

# 6. Investment Rate
test_df["Investment_Rate"] = test_df["Amount_invested_monthly"] / (test_df["Monthly_Inhand_Salary"] + 1)

# 7. Credit History in Total Months
if "Credit_History_Age" in test_df.columns:
    extracted = test_df["Credit_History_Age"].str.extract(r'(\d+)\s+Years.*?(\d+)\s+Month')
    test_df["Credit_History_Years"] = pd.to_numeric(extracted[0], errors='coerce')
    test_df["Credit_History_Months"] = pd.to_numeric(extracted[1], errors='coerce')
    test_df["Credit_History_Total_Months"] = test_df["Credit_History_Years"] * 12 + test_df["Credit_History_Months"]
    test_df.drop(["Credit_History_Years", "Credit_History_Months"], axis=1, inplace=True)

# 8. Payment Behaviour - High Spent or not
test_df["PB_HighSpent"] = test_df["Payment_Behaviour"].apply(lambda x: "High" if "High_spent" in str(x) else "Low")

# 9. Payment Behaviour - Over EMI or not
test_df["PB_OverEMI"] = test_df["Payment_Behaviour"].apply(lambda x: "Over" if "More than EMI" in str(x) else "Under")

# 10. Age Bucketing
test_df["Age_Bucket"] = pd.cut(test_df["Age"], bins=[0, 25, 40, 60, 100], labels=["Youth", "Adult", "Mid-Age", "Senior"])

# 11. Net Monthly Balance
test_df["Net_Monthly_Balance"] = test_df["Monthly_Inhand_Salary"] - test_df["Total_EMI_per_month"]

# 12. Unknown Credit Mix
test_df["Unknown_CreditMix"] = test_df["Credit_Mix"].apply(lambda x: 1 if str(x).strip().upper() == "NM" else 0)
````

# **Project Limitations**

Despite the promising results, the project encountered a few practical and computational challenges:

- **Hyperparameter Tuning Overhead:** Although hyperparameter tuning improved model performance, it significantly increased computation time. RandomizedSearchCV and GridSearchCV both required substantial resources, especially on larger parameter grids.
- **Cross-Validation Cost:** Cross-validation enhanced model reliability, but it increased training time considerably. A single run using 5-fold cross-validation took approximately 5 minutes, which could become a bottleneck with larger datasets or more complex models.
- **Encoding Challenges:** Binary encoding was initially explored to compress high-cardinality categorical variables, but it was ultimately avoided. It conflicted with the project objective of maintaining interpretability and traceability in model inputs.
- **Model Interpretability Constraints (XGBoost):** XGBoost, while powerful, provides less transparent feature importances compared to Random Forest and is harder to explain to stakeholders without SHAP or LIME visualizations, which were outside the current project scope.
- **No Class Weighting or Imbalance Adjustment:** The current approach treated all classes equally. If class imbalance exists (e.g., fewer "Good" or "Poor" labels), it may bias predictions toward the dominant class. Future work should consider SMOTE or class-weight adjustments.
